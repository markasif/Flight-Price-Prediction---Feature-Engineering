```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
dataa = pd.read_excel('flight_price.xlsx')
dataa.head()
```

```python
dataa.info()
```

```python
dataa.describe()
```

```python
dataa["day"] = dataa["Date_of_Journey"].str.split('/').str[0]
dataa["month"] = dataa['Date_of_Journey'].str.split('/').str[1]
dataa["year"] = dataa['Date_of_Journey'].str.split('/').str[2]
```

```python
dataa
```

```python
dataa.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
dataa
```

```python
dataa.info()
```

```python
dataa['day'] = dataa['day'].astype(int)
dataa['month'] = dataa['month'].astype(int)
dataa['year'] = dataa['year'].astype(int)
```

```python
dataa
```

```python
dataa["Arrival_Time"] = dataa["Arrival_Time"].apply(lambda x:x.split(' ')[0])
```

```python
dataa
```

```python
dataa["Arrival_hour"] = dataa["Arrival_Time"].str.split(':').str[0]
dataa["Arrival_minute"] = dataa ['Arrival_Time'].str.split(':').str[1]
```

```python
dataa.drop("Arrival_Time",axis=1,inplace=True)
```

```python
dataa.info()
```

```python
dataa["Arrival_hour"] = dataa["Arrival_hour"].astype(int)
dataa["Arrival_minute"] = dataa["Arrival_minute"].astype(int)
```

```python
dataa.head()
```

```python
dataa["Dep_hour"] = dataa["Dep_Time"].str.split(':').str[0]
dataa["Dep_minute"] = dataa ['Dep_Time'].str.split(':').str[1]
```

```python
dataa.drop('Dep_Time',axis=1,inplace=True)
```

```python
dataa['Dep_hour'] = dataa['Dep_hour'].astype(int)
dataa['Dep_minute'] = dataa['Dep_minute'].astype(int)
```

```python
dataa.head()
```

```python
dataa['Total_Stops'].unique()
```

```python
dataa[dataa["Total_Stops"].isnull()]
```

```python
dataa['Total_Stops'].mode()
```

```python
dataa['Total_Stops'] = dataa['Total_Stops'].map({'non-stop':0, '1 stop': 1, '2 stops': 2, '3 stops' : 3, '4 stops': 4, np.nan: 1})
```

```python
dataa.head()
```

```python
dataa['Total_Stops'].isnull().sum()
```

```python
dataa.drop('Route', axis=1,inplace=True)
```

```python
dataa['Duration_hour'] = dataa['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
dataa['Duration_minutes'] = dataa['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
dataa['Duration_minutes'].isnull().sum()
```

```python
dataa['Duration_hour'].unique()
```

```python
dataa['Duration_hour'] = dataa['Duration_hour'].replace('5m',5)
```

```python
dataa['Duration_minutes'].unique()
```

```python
dataa.info()
```

```python
dataa['Duration_hour'] = dataa['Duration_hour'].astype(int)
```

```python
dataa['Duration_minutes'].isnull().sum()
```

```python
dataa['Duration_minutes'].isnull().mean() * 100
```

```python
dataa['Duration_minutes'] = pd.to_numeric(dataa['Duration_minutes'], errors='coerce')
dataa['Duration_minutes'] = dataa['Duration_minutes'].astype(int)
```

```python
dataa['Duration_minutes'] = dataa['Duration_minutes'].fillna(dataa['Duration_minutes'].mean())
```

```python
dataa.info()
```

```python
dataa.head()
```

```python
dataa.drop("Duration",axis=1,inplace=True)
```
